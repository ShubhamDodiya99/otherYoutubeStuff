```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
y=x*x
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r*--', markersize=8)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g*-')
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
x = np.arange(1,11)
y = x*x*x
plt.title('matplotlit demo')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x,y)
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y)
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')

plt.show()
```

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.show()
```

```python
a = np.array([22,86,55,44,34,54,12,34,56,78,99,87,23,12,13,20,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

plt.boxplot(data,vert=True, patch_artist=True);
```
